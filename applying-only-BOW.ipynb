```python
# Applying Bag-of-Words on 'question1' and 'question2'
# columns and 'is_duplicate' is the output column , 
# Without applying any Preprocessing and Feature Engineering and 
# checking the Accuracy Score.
# Applyed RandomForest Classifier
```

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
df.head(3)
df.shape
```

```python
# Taking only 30000 Rows
new_df=df.sample(30000)
new_df.isnull().sum()
```

```python
new_df.duplicated().sum()
```

```python
# Creating new Dataframe using only 'question1' and 'question2' columns
ques_df=new_df[['question1','question2']]
# ques_df.head()
```

```python
# Appling BOW
from sklearn.feature_extraction.text import CountVectorizer
# Fill NaNs with empty string in both columns
ques_df['question1'] = ques_df['question1'].fillna('')
ques_df['question2'] = ques_df['question2'].fillna('')
questions = list(ques_df['question1']) + list(ques_df['question2'])
cv = CountVectorizer(max_features=3000)
q1_arr,q2_arr=np.vsplit(cv.fit_transform(questions).toarray(),2)

```

```python
# Convert both arrays into DataFrame and Concatenate
temp_df1=pd.DataFrame(q1_arr,index=ques_df.index)
temp_df2=pd.DataFrame(q2_arr,index=ques_df.index)
temp_df=pd.concat([temp_df1,temp_df2],axis=1)
temp_df.shape
```

```python
# Matrix Reppresentation
temp_df
```

```python
# Add 'is_duplicate' column to temp_df dataframe
temp_df['is_duplicate']=new_df['is_duplicate']
temp_df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(temp_df.iloc[:,0:-1].values,temp_df.iloc[:,-1].values,test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
y_pred=xgb.predict(x_test)
accuracy_score(y_test,y_pred)
```
