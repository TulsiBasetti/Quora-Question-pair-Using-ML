```python
# Feature Engineering

new_df['q1_len']=new_df['question1'].str.len()
new_df['q2_len']=new_df['question2'].str.len()
new_df.head()
```

```python
new_df['q1_nums_words']=new_df['question1'].apply(lambda row:len(row.split(" ")))
new_df['q2_nums_words']=new_df['question2'].apply(lambda row:len(row.split(" ")))
new_df.head()
```

```python
# Get common words in both questions
# Strip() -> remove leading and trailing whitespace from a string
# Set-> No Duplicate words are present
def common_words(row):
    w1=set(map(lambda word:word.lower().strip(),row['question1'].split(" ")))
    w2=set(map(lambda word:word.lower().strip(),row['question2'].split(" ")))
    return len(w1&w2)
```

```python
new_df['word_common']=new_df.apply(common_words,axis=1)
new_df.head()
```

```python
# Total words in both questions

def total_words(row):
    w1=set(map(lambda word:word.lower().strip(),row['question1'].split(" ")))
    w2=set(map(lambda word:word.lower().strip(),row['question2'].split(" ")))
    return len(w1)+len(w2)
```

```python
new_df['word_total']=new_df.apply(total_words,axis=1)
new_df.head()
```

```python
new_df['word_share']=round(new_df['word_common']/new_df['word_total'],2)
new_df.head()
```

```python
# Applying BOW
ques_df = new_df[['question1','question2']]
ques_df.head()
```

```python
final_df = new_df.drop(columns=['id','qid1','qid2','question1','question2'])
print(final_df.shape)
final_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
questions = list(ques_df['question1']) + list(ques_df['question2'])
cv = CountVectorizer(max_features=3000)
q1_arr, q2_arr = np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python

# Convert both arrays into DataFrame and Concatenate
temp_df1=pd.DataFrame(q1_arr,index=ques_df.index)
temp_df2=pd.DataFrame(q2_arr,index=ques_df.index)
temp_df=pd.concat([temp_df1,temp_df2],axis=1)
temp_df.shape
```

```python
final_df = pd.concat([final_df, temp_df], axis=1)
print(final_df.shape)
final_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(final_df.iloc[:,1:].values,final_df.iloc[:,0].values,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
accuracy_score(y_test,y_pred)
```
