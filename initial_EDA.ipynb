```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
df.head(3)
df.shape
```

```python
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Check for the distribution of duplicate question pair and non-duplicate question pair
# Bar Graph Implementation
counts=df['is_duplicate'].value_counts()
print(counts)
print((df['is_duplicate'].value_counts()/df['is_duplicate'].count())*100)
#df['is_duplicate'].value_counts().plot(kind='bar')
plt.figure(figsize=(6,4))
counts.plot(kind='bar', color=['skyblue', 'lightcoral'], edgecolor='black')
plt.title('Distribution of Duplicate vs Non-Duplicate Questions')
plt.xlabel('Is Duplicate (0 = No, 1 = Yes)')
plt.ylabel('Number of Question Pairs')
```

```python
# Number of Repeated Questions in the Dataset
qid=pd.Series(df['qid1'].tolist()+df['qid2'].tolist())
print('Number of Unique Questions',np.unique(qid).shape[0])
x=qid.value_counts()>1 # repetation of each question id
print('Number of Questions Getting Repeded',x[x].shape[0]) #selects only repeded questions
```

```python
# Histogram representaion of Repeated questions
plt.hist(qid.value_counts().values, bins=160)
plt.yscale('log')
plt.xlabel('Number of times question appears')
plt.ylabel('Number of questions')
plt.title('Distribution of Question Repetition in Dataset (Log Scale)')
plt.grid(True)
plt.show()
```
